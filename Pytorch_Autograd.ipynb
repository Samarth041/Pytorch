# 🤖 What is Autograd in PyTorch?

## Simple Definition

**Autograd** is PyTorch's **automatic differentiation engine**.

It automatically computes gradients required for training neural networks.

Instead of manually calculating derivatives, PyTorch does it for you using **backpropagation**.

---

## What Does Autograd Do?

Autograd:

- Tracks all mathematical operations performed on tensors
- Builds a computation graph automatically
- Uses the **Chain Rule** from calculus
- Computes gradients during backpropagation
- Helps optimize neural network parameters

## Why Do We Need Autograd?

Training a neural network requires calculating gradients of the loss function with respect to model parameters.

For a large neural network containing millions of parameters, calculating these gradients manually is impossible.

Autograd solves this problem by:

1. Recording every operation
2. Creating a computation graph
3. Automatically computing gradients

```python
import torch

# Inputs
x = torch.tensor(6.7)  # Input feature
y = torch.tensor(0.0)  # True label (binary)

#initial value
w = torch.tensor(1.0)  # Weight
b = torch.tensor(0.0)  # Bias
```

```python
# Binary Cross-Entropy Loss for scalar
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8  # To prevent log(0)
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
```

```python
# Forward pass
z = w * x + b  # Weighted sum (linear part)
y_pred = torch.sigmoid(z)  # Predicted probability

# Compute binary cross-entropy loss
loss = binary_cross_entropy_loss(y_pred, y)
```

```python
loss
```

```python
# Derivatives:
# 1. dL/d(y_pred): Loss with respect to the prediction (y_pred)
dloss_dy_pred = (y_pred - y)/(y_pred*(1-y_pred))

# 2. dy_pred/dz: Prediction (y_pred) with respect to z (sigmoid derivative)
dy_pred_dz = y_pred * (1 - y_pred)

# 3. dz/dw and dz/db: z with respect to w and b
dz_dw = x  # dz/dw = x
dz_db = 1  # dz/db = 1 (bias contributes directly to z)

dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f"Manual Gradient of loss w.r.t weight (dw): {dL_dw}")
print(f"Manual Gradient of loss w.r.t bias (db): {dL_db}")
```
