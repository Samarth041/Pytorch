## Defining the model


````python
#create model

import torch
import torch.nn as nn
````

````python
import torch.nn as nn


class MySimpleNN(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):

    out = self.linear(features)
    out = self.sigmoid(out)

    return out
````

Important Parameters

````python
learning_rate = 0.1
epochs = 25
````

````python
# define loss function
loss_function = nn.BCELoss()
````

# Pytorch NN module

````python
# create model class
import torch
import torch.nn as nn

class Model(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )

  def forward(self, features):

    out = self.network(features)

    return out
````

````python
# create model
model = MySimpleNN(X_train_tensor.shape[1])

# define loop
for epoch in range(epochs):

    # forward pass
    y_pred = model(X_train_tensor)

    # calculate loss
    loss = loss_function(y_pred, y_train_tensor.view(-1,1))

    # clear old gradients
    model.linear.weight.grad = None
    model.linear.bias.grad = None

    # backward pass
    loss.backward()

    # manually update parameters
    with torch.no_grad():
        model.linear.weight -= learning_rate * model.linear.weight.grad
        model.linear.bias -= learning_rate * model.linear.bias.grad

    # print loss
    print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
````

## What is `torch.optim`?

`torch.optim` is the optimization module in PyTorch used to update the parameters (weights and biases) of a neural network during training.

It provides several built-in optimization algorithms, such as:

- **SGD (Stochastic Gradient Descent)**
- **Momentum**
- **RMSProp**
- **Adam**
- **Adadelta**
- **AdamW**
- **Adagrad**

### Why use `torch.optim`?

Instead of manually updating parameters:

```python
w = w - lr * w.grad
b = b - lr * b.grad
```

PyTorch can handle it automatically:

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

optimizer.step()      # Update parameters
optimizer.zero_grad() # Clear gradients
```

### Advantages

- Automatic parameter updates
- Supports many optimization algorithms
- Cleaner and more readable code
- Additional features like:
  - Learning rate scheduling
  - Weight decay (L2 regularization)
  - Momentum-based updates

### Example

```python
import torch.optim as optim

optimizer = optim.Adam(model.parameters(), lr=0.001)

for epoch in range(epochs):
    optimizer.zero_grad()
    
    y_pred = model(X_train)
    loss = loss_function(y_pred, y_train)
    
    loss.backward()
    optimizer.step()
```

### Training Workflow

```text
Forward Pass
      ↓
Calculate Loss
      ↓
Backward Pass (loss.backward())
      ↓
optimizer.step()
      ↓
optimizer.zero_grad()
```

````python
# create model
model = MySimpleNN(X_train_tensor.shape[1])

# define optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# define loop
for epoch in range(epochs):

  # forward pass
  y_pred = model(X_train_tensor) #internaly call lmodel.forward

  # loss calculate
  loss = loss_function(y_pred, y_train_tensor.view(-1,1))

  # clear gradients
  optimizer.zero_grad()

  # backward pass
  loss.backward()

  # parameters update
  optimizer.step()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
````
